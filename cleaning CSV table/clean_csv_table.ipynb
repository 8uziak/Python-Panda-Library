```python
df = pd.read_csv('BCL_TaskCreation.csv',delimiter=';')
df
# I used a delimiter to properly separate the columns
# now table has 3 columns with headers dataStream, rowCount and probingTimestamp
```

```python
df[['Date','Time']] = df['probingTimestamp'].str.split(' ', expand=True)
df.drop('probingTimestamp', inplace=True, axis=1)
df

# I split the probingTimestamp column containing date and time together into two separate columns:
# 'Date' and 'Time'
# I removed the original column (probingTimestamp) to avoid repeating the same data
# table has 4 columns and 160735 rows
```

```python
df['dataStream'].unique()
# I checked to see if it was worth leaving the 'dataStream' column 
# I used a function to check the number of unique values in a cell
# the output told me that the cell named 'BCL_TaskCreation' repeats 160734 times
# so from the beginning (without header of course) to the last index the cells are the same
```

```python
# I decided to remove the 'dataStream' column due to its worthless nature in this table

df.drop('dataStream', inplace=True, axis=1)
df
```

```python
df.sort_values(by=['Time'], inplace=True)
df
# there is no specific hourly time, from the beginning of the day to the end of the day 
```

```python
df.sort_values(by=['Date','Time'], inplace=True)
df
# I noticed that the data in the table is not sorted in any way
# you can see above in the 'Date' column in rows 1, 2 and 3 that the dates are not in sequence
# I came to the conclusion to sort the table by cells from the earliest date to the oldest
# and from earliest time to latest time
```
